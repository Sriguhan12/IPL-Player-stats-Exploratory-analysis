# Categorizing players based on age

- Young players - age less than 25

- Mid aged players - age between 25 and 30

- Senior players - more than 30

```python
data.loc[data['Age'] <= '25 Years','age_category']='Young'
data.loc[(data['Age'] > '25 Years') & (data['Age'] <= '30 Years') ,'age_category']='Mid aged player'
data.loc[data['Age'] > '30 Years','age_category']='Senior Player'
data_age = data[['Name','age_category']].value_counts('age_category')
data_age
```

# Best emerging player in IPL

Conditions:
 1. Age should be less than 25
 2. Should have played atleast 15 matches
 3. Emerging player can be of any type in the sense he can be either batsman,bowler,all rounder,wicketkeeper

```python
df_emerge_bowl = data[(data.Age <= '25 years') & (data.MatchPlayed >= 15) & (data['Type'] == 'Bowler ')]
df_bowl1 = df_emerge_bowl.groupby(['Full Name','Wickets']).max().sort_values('Wickets',ascending = False).head(1)
df_bowl1
```

```python
df_emerge_bat = data[(data.Age <= '25 years') & (data.MatchPlayed >= 15) & (data['Type'] == 'Batsman ')]
df_bat1 = df_emerge_bat.groupby(['Full Name','RunsScored']).max().sort_values('RunsScored',ascending = False).head(1)
df_bat1
```

```python
df_emerge_All = data[(data.Age <= '25 years') & (data.MatchPlayed >= 15) & (data['Type'] == 'All-Rounder ')]
df_all1 = df_emerge_All.groupby(['Full Name','RunsScored']).max().sort_values('Wickets',ascending = False).head(1)
df_all1
```

```python
df_emerge_WK = data[(data.Age <= '25 years') & (data.MatchPlayed >= 15) & (data['Type'] == 'Wicket-Keeper ')]
df_wk1 = df_emerge_WK.groupby(['Full Name','RunsScored']).max().sort_values('RunsScored',ascending = False).head(1)
df_wk1
```

```python
df_emerge_result = df_bowl1.append([df_bat1 , df_all1 , df_wk1])
df_emerge_result
```

These four players are in the final shortlisted for emerging player and the award will be given based on votes from fans.

# Plot to compare the batting style of indian players and foreign players

 Condtions:
 
1. This plot will show how indian and foreign players score runs, in the sense whether they score more by hitting boundaries or they tend to take more 1s ,2s and 3s
2. lets consider top 20 players

# Indian player

```python
plt.figure(figsize = (10,5))
data_ind = data[data['National Side'] == 'India']
data_ind1 = data_ind[['Name','4s','6s','RunsScored']].sort_values('RunsScored',ascending = False).head(20)
plt.subplot(1,2,1)
plt.scatter(data = data_ind1,x='Name',y='4s')
plt.title('Boundaries')
plt.xlabel("Players")
plt.ylabel("No. of 4s")
plt.xticks([])
plt.subplot(1,2,2)
plt.scatter(data = data_ind1,x='Name',y='6s')
plt.title("Sixers")
plt.xlabel("Players")
plt.ylabel("No. of 6s")
plt.xticks([])
plt.subplots_adjust(left=0.01)
```

# Overseas player

```python
plt.figure(figsize = (10,5))
data_ind = data[data['National Side'] != 'India']
data_ind1 = data_ind[['Name','4s','6s','RunsScored']].sort_values('RunsScored',ascending = False).head(20)
plt.subplot(1,2,1)
plt.scatter(data = data_ind1,x='Name',y='4s')
plt.title('Boundaries')
plt.xlabel("Players")
plt.ylabel("No. of 4s")
plt.xticks([])
plt.subplot(1,2,2)
plt.scatter(data = data_ind1,x='Name',y='6s')
plt.title("Sixers")
plt.xlabel("Players")
plt.ylabel("No. of 6s")
plt.xticks([])
plt.subplots_adjust(left=0.01)
```
